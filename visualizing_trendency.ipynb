```python
import torch 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
list_c4synth_perplexity = [21.12, 18.42, 19.01] 
list_c4_perplexity = [50.35, 44.17, 46.2] 
list_pg19_perplexity = [76.32, 123.56, 108.89] 
```

```python
checkpoint_list = [1000, 2000, 3000] 
plt.plot(checkpoint_list, list_c4synth_perplexity, label='c4synth', marker='o') 
plt.plot(checkpoint_list, list_c4_perplexity, label='c4', marker = "o") 
plt.plot(checkpoint_list, list_pg19_perplexity, label='pg19', marker = "o") 
plt.xlabel("step count") 
plt.ylabel("perplexity") 
plt.grid() 
plt.legend() 
plt.show() 
```

```python
ce_loss_c4synth = [3.05, 2.91, 2.94] 
ce_loss_c4 = [3.91, 3.78, 3.83] 
ce_loss_pg19 = [4.81, 4.69, 4.74] 
plt.plot(checkpoint_list, ce_loss_c4synth, label='c4synth', marker='o') 
plt.plot(checkpoint_list, ce_loss_c4, label='c4', marker = "o") 
plt.plot(checkpoint_list, ce_loss_pg19, label='pg19', marker = "o") 
plt.xlabel("step count") 
plt.ylabel("cross entropy loss") 
plt.grid() 
plt.legend() 
plt.show() 
```

```python
kernelsize7_position = [0.767, 0.756, 0.702, 0.685, 0.704, 0.697, 0.702] 
kernelsize2_position = [0.801, 0.736] 
kernelsize1_position = 0.802 
finetuned = 0.6954 
plain = 0.6362 
```

```python
plt.plot(list(range(1, 8)), kernelsize7_position, label = 'kernelsize7', marker = 'o', color = 'blue') 
plt.plot(list(range(1, 3)), kernelsize2_position, label = 'kernelsize2', marker = 'x', color = 'green') 
plt.plot(1, kernelsize1_position, label = 'kernelsize1', marker = 'p', color = 'red') 

# ploting horizontal line 
plt.axhline(y = finetuned, color = 'black', linestyle = "--", label = 'small model finetuned') 
plt.axhline(y = plain, color = 'green', linestyle = '--', label = 'small model plain') 
plt.annotate('small model finetuned', xy = (7, finetuned), xytext = (7.5, finetuned), arrowprops = dict(facecolor = 'black', arrowstyle = '->')) 
plt.annotate('small model plain', xy = (7, plain), xytext = (7.5, plain), arrowprops = dict(facecolor = 'green', arrowstyle = '->')) 

plt.xlabel("Position After the Condensed Token") 
plt.ylabel("Acceptance Rate") 
plt.grid() 
plt.legend() 

plt.show() 
```
